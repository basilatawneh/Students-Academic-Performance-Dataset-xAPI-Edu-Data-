```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns; 
import statsmodels.api as sm
import random
data=pd.read_csv('xAPI-Edu-Data.csv')
data.head()
```

```python
data.describe()
```

# check the null vaules  

```python
data.isnull().sum()
```

# convert the string values to numbers and cleaning the data

```python
def cleanstage(x):
    if x=='lowerlevel' or x=='L':
        return 0
    elif x=="MiddleSchool"or x=='M':
        return 1
    else:
        return 2
data.StageID=data.StageID.apply(lambda y:cleanstage(y))
data['gender']=data.gender.apply(lambda x: 0 if x=='M' else 1)
data.ParentschoolSatisfaction=data.ParentschoolSatisfaction.apply(lambda x: 1 if x=='Good' else 0)
data.StudentAbsenceDays=data.StudentAbsenceDays.apply(lambda x: 0 if x=="Under-7" else 1)
data.Semester=data.Semester.apply(lambda x: 0 if x=='F' else 1)# 0=F - 1=S
data.GradeID=data.GradeID.apply(lambda x: int(x[2:]))
data.Relation=data.Relation.apply(lambda x: 0 if x=='Father' else 1)
data.ParentAnsweringSurvey=data.ParentAnsweringSurvey.apply(lambda x: 1 if x=='Yes' else 0)
data.Class=data.Class.apply(lambda x:cleanstage(x))
data.head()
```

# find the correlation 

```python
corr = data.corr()
corr.style.background_gradient(cmap='coolwarm').set_precision(2)
```

```python

d=data.groupby(['Class','StudentAbsenceDays'])['StudentAbsenceDays'].count()
#plt.bar(data.Class,d)

n_groups = 3
print(d)
x = (d[0][0],d[1][0], d[2][0])
y = (d[0][1], d[1][1], d[2][1])

# create plot
fig, ax = plt.subplots()
index = np.arange(n_groups)
bar_width = 0.35
opacity = 0.8

rects1 = plt.bar(index, x, bar_width,
alpha=opacity,
color='b',
label='Under-7')

rects2 = plt.bar(index + bar_width, y, bar_width,
alpha=opacity,
color='g',
label='above-7')

plt.xlabel('total grade (class)')
plt.ylabel('Absence Days')
plt.title('relation between the grade and absent days')
plt.xticks(index + bar_width, ('Low', 'Middle', 'High'))
plt.legend()

plt.tight_layout()
plt.show()
```

# there is relation between the grade of the student and how many absent days
# if the grade is high then the absent days is low
